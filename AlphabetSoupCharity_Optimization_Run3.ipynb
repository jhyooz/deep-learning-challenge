## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_new_df = application_df.drop(columns=["EIN"], axis=1)
application_new_df
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()

# Display unique values
unique_values
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Display value counts
application_type_counts
```

```python
# DUPE OF APPLICATION TYPES...UPDATE AS NEEDED
# Choose a cutoff value and create a list of names to be replaced
cutoff_value = 50
# Create a list of Names to be replaced
application_types_to_replace = application_new_df['NAME'].value_counts()[application_new_df['NAME'].value_counts() < cutoff_value].index.tolist()
# Replace in dataframe
for app in application_types_to_replace:
    application_new_df['NAME'] = application_new_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_new_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Set the cutoff value
cutoff = 100

application_types_to_replace = application_type_counts[application_type_counts < cutoff].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Display value counts
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Filter classification_counts for counts > 1
classification_counts = classification_counts[classification_counts > 1]

# Display value counts greater than 1
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# Set the cutoff value
cutoff = 125

classifications_to_replace = application_new_df['CLASSIFICATION'].value_counts()[application_new_df['CLASSIFICATION'].value_counts() < cutoff_value].index.tolist()
# Replace in dataframe
for cls in classifications_to_replace:
    application_new_df['CLASSIFICATION'] = application_new_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_new_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
data_encoded = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
X = data_encoded.drop(["IS_SUCCESSFUL"], axis='columns').values
y = data_encoded["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()
number_input_features = len(X_train[0])
hidden_layer1 = 100
hidden_layer2 = 30
hidden_layer3 = 10

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer1, activation="relu", input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
fit_model
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)

# Print the model's loss and accuracy
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
